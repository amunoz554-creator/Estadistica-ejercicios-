```python
# Análisis de Vacunación (muestra)

# 1) Instalación de librerías adicionales
!pip install -q missingno

# 2) Importes
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
from IPython.display import display

plt.rcParams['figure.figsize'] = (10,6)
sns.set_style('whitegrid')

# 3) Subir archivo CSV
print("archivo vacunacion_sample.csv")
uploaded = files.upload()

df = pd.read_csv("vacunacion_sample.csv")
print("Dimensiones:", df.shape)
display(df.head())

# 4) Valores nulos
print("Conteo de nulos por columna:")
display(df.isnull().sum())
msno.matrix(df, figsize=(12,4))
plt.show()

# 5) Selección de variables
cat1, cat2 = "Departamento", "Vacuna"
num1, num2 = "Dosis_aplicadas", "Fecha"

print("Categoricas:", cat1, ",", cat2)
print("Numéricas:", num1, ",", num2)

# 6) Gráficas categóricas
def plot_categoria(col, top_n=10):
    vals = df[col].value_counts().head(top_n)
    plt.figure(figsize=(10,5))
    sns.barplot(x=vals.values, y=vals.index)
    plt.title(f"Conteo por {col}")
    plt.show()

    plt.figure(figsize=(6,6))
    vals.plot.pie(autopct='%1.1f%%')
    plt.title(f"Distribución {col}")
    plt.ylabel("")
    plt.show()

plot_categoria(cat1)
plot_categoria(cat2)

# 7) Gráficas numéricas
df['Fecha'] = pd.to_datetime(df['Fecha'])
df['Año'] = df['Fecha'].dt.year

def plot_numerica(col):
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f"Histograma {col}")

    plt.subplot(1,2,2)
    sns.boxplot(x=df[col].dropna())
    plt.title(f"Boxplot {col}")

    plt.tight_layout()
    plt.show()

plot_numerica(num1)
plot_numerica("Año")

# 8) Outliers con IQR
def remove_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    mask = (data[col] >= lower) & (data[col] <= upper)
    return data.loc[mask], (lower, upper)

df_clean, limits = remove_outliers_iqr(df, "Dosis_aplicadas")
print("Original:", df.shape, " Limpio:", df_clean.shape)
print("Límites aplicados:", limits)

# 9) Guardar dataset limpio
df_clean.to_csv("vacunacion_sample_clean.csv", index=False)
print("✅ Dataset limpio guardado como vacunacion_sample_clean.csv")

```
