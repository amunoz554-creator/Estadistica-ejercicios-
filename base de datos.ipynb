```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar archivo CSV
df = pd.read_csv("Escuelas_deportivas.csv")

# Información básica
print("Información del dataset")
print(df.info())
print("Primeras filas del dataset")
print(df.head())
print("Resumen estadístico del dataset")
print(df.describe())
```

```python
# Valores faltantes
print("\nValores faltantes por columna:")
print(df.isnull().sum())

# Visualización de nulos
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes")
plt.show()
```

```python
df_selected = df[['Año', 'Barrio', 'Parque o escenario', 'Disciplinas', 'Edades', 'Dias', 'Hora de clase']]
display(df_selected.head())
```

```python
# Valores faltantes
print("\nValores faltantes por columna:")
print(df_selected.isnull().sum())

# Visualización de nulos
sns.heatmap(df_selected.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes")
plt.show()
```

```python
# Diagrama de torta para la variable 'Barrio'
barrio_counts = df['Barrio'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(barrio_counts, labels=barrio_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Escuelas Deportivas por Barrio')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Diagrama de barras para la variable 'Barrio'
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Barrio', order=barrio_counts.index, palette='viridis')
plt.title('Número de Escuelas Deportivas por Barrio')
plt.xlabel('Número de Escuelas')
plt.ylabel('Barrio')
plt.show()
```

```python
# Diagrama de torta para la variable 'Parque o escenario'
parque_counts = df['Parque o escenario'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(parque_counts, labels=parque_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Escuelas Deportivas por Parque o Escenario')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Diagrama de barras para la variable 'Parque o escenario'
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Parque o escenario', order=parque_counts.index, palette='viridis')
plt.title('Número de Escuelas Deportivas por Parque o Escenario')
plt.xlabel('Número de Escuelas')
plt.ylabel('Parque o Escenario')
plt.show()
```

```python
# Diagrama de torta para la variable 'Disciplinas'
disciplina_counts = df['Disciplinas'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(disciplina_counts, labels=disciplina_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Escuelas Deportivas por Disciplina')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Diagrama de barras para la variable 'Disciplinas'
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Disciplinas', order=disciplina_counts.index, palette='viridis')
plt.title('Número de Escuelas Deportivas por Disciplina')
plt.xlabel('Número de Escuelas')
plt.ylabel('Disciplina')
plt.show()
```

```python
# Select only numerical columns for IQR calculation
df_numeric = df.select_dtypes(include=np.number)

# Calculate the IQR for numerical columns
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers in numerical columns
outliers = df_numeric[(df_numeric < lower_bound) | (df_numeric > upper_bound)]

# Remove rows with outliers based on numerical columns
df_cleaned = df[~((df_numeric < lower_bound) | (df_numeric > upper_bound)).any(axis=1)]

print("Original DataFrame shape:", df.shape)
print("Cleaned DataFrame shape:", df_cleaned.shape)

# Display the first few rows of the cleaned DataFrame
display(df_cleaned.head())
```

```python
# Histogram for the 'Año' column in the cleaned DataFrame
plt.figure(figsize=(8, 6))
sns.histplot(data=df_cleaned, x='Año', kde=True)
plt.title('Distribución del Año en los Datos Limpios')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.show()
```
